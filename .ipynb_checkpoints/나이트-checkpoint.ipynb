```python
from collections import deque

def solution(n, startRow, startCol, endRow, endCol):
    dx = [-1, -2, -2, -1, 1, 2, 2, 1]
    dy = [2, 1, -1, -2, -2, -1, 1, 2]
    
    def bfs(startRow, startCol, endRow, endCol):
        q = deque()
        q.append([startRow, startCol])
        chess[startRow][startCol] = 1
        while q:
            a, b = q.popleft()
            if a == endRow and b == endCol:
                return(chess[endRow][endCol] -1)
            
            for i in range(8):
                x = a + dx[i]
                y = b + dy[i]
                if 0 <= x < n and 0 <= y < n and chess[x][y] == 0:
                    q.append([x, y])
                    chess[x][y] = chess[a][b] + 1
    chess = [[0] * n for i in range(n)]
    
    return bfs(startRow, startCol, endRow, endCol)
```

```python
d_x = [1,1,2,2,-1,-1,-2,-2]
d_y = [2,-2,1,-1,2,-2,1,-1]
```

```python
def BFS(start_x ,start_y , end_x, end_y):
    q= deque()
    q.append([start_x ,start_y])
    map_[start_x][start_y] = 1
    
    while q:
        pop_x , pop_y = q.popleft()
        if pop_x == end_x and pop_y == end_y:
            return(map_[end_x][end_y]-1)
        
        for i in range(len(d_x)):
            next_x  = pop_x + d_x[i]
            next_y  = pop_y + d_y[i]
        
            if 0 <= next_x < n and 0 <= next_y < n and map_[next_x][next_y] == 0:
                q.append([next_x, next_y])
                map_[next_x][next_y] =map_[pop_x][pop_y]+1
    
 
```
