```python
class DList:
    class Node:
        def __init__(self , prev, item, next_item ):
            self.prev = prev
            self.item = item
            self.next_item = next_item
    
    def __init__(self):
        '''
        Initialize head and tail
        Start to set head None
        Set tail's head -> self.head
        Set size -> 0 
        '''
        self.head  = self.Node(None, None, None) # init head, tail ,item None
        self.tail  = self.Node(self.head  , None, None )
        self.head.next_item = self.tail
        self.size = 0
        
    def get_size(self):
        '''
        Return self.size
        '''
        
        return self.size
    
    def is_empty(self):
        '''
        Return Ture/False
        '''
        return self.size == 0
    
    def add(self , pos , item):
        '''
        Add item to pos
        p is self.head and Pointer
        Move pointer to pos in iteration
        
        1 <=> 2 <=> ...
        t     p
        
        1 <=> 2 <=> 3
        t     n     P
        
        '''
        
        p = self.head
        for i in range(pos):
            p = p.next_item
        t = p.prev
        n = self.Node(t, item, p)
        t.next_item = n
        p.prev = n
        self.size += 1
        
    def delete(self, pos):
        if self.size < pos:
            return
        
        p =  self.head
        
        for i in range(pos):
            p = p.next_item
        
        t = p.prev
        q = p.next_item
        t.next_item = q
        q.prev = t
        self.size -=1
    
    
    def get(self, pos):
        if self.size < pos:
            return "Invalid position"
        
        p =  self.head
        for i in range(pos):
            p = p.next_item
        
        return p.item
    
    def print(self):
        if self.is_empty():
            print("List is sempty.")
        
        else:
            p = self.head.next_item
            while p!=  self.tail:
                if p.next_item != self.tail:
                    print(p.item ,end = "")
                
                else:
                    print(p.item)
                    
                p = p.next_item
            
```

```python
def ADT():
    num = int(input("Number of operations : "))
    List = DList()
    
    for i in range(num):
        inputs = input("Input : ")
        operation  = inputs[0]
        if operation == "A":
            pos = int(inputs[2:-1])
            alphabet = inputs[-1]
            List.add(pos,alphabet)
        elif operation == "D":
            pos =  int(inputs[2:])
            List.delete(pos)
        elif operation == "G":
            pos = int(inputs[2:])
            print(List.get(pos))
        elif operation == 'P':
            List.print()
    
```
