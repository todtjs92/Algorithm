```python
from itertools import combinations
def solution():
    count = int(input())
    sum_li = []
    mul_li = []
    for _ in range(count):
        sum_num , mul_num = list(map(int,input().split()))
        sum_li.append(sum_num)
        mul_li.append(mul_num)
    minimum =  1000000000
    for c in range(1,count+1):
        t1 = combinations(sum_li,c)
        t2 =combinations(mul_li,c)
        for a,b in zip(t1,t2):
            sum_a = sum(a)
            list_b = list(b)
            mul_b = 1
            for i in range(len(b)):
                mul_b = mul_b * list_b[i]
            temp = abs(mul_b - sum_a)
            minimum = min(temp, minimum)
            if minimum == 0:
                print(minimum)
                return 
    print(minimum)
```

### 상자고르기

```python
AABBB
BBBBA
BBBBA
BBBBB
BBBBB
```

```python
from copy import deepcopy
map_ = []
for _ in range(5):
    string = input()
    temp = []
    for s in string:
        temp.append(s)
    map_.append(temp)


answer = 0


def BFS(i, j, count_all, count_a, visit):
    
    global answer
    count_all += 1

    if map_[i][j] == 'A':
        count_a += 1
    new_visit = deepcopy(visit)
    new_visit.append([i, j])
    if count_all == 7:
        if count_a >= 4:
            answer += 1
            print(new_visit)
            return
        else:
            return

    dx = [1, -1, 0, 0]
    dy = [0, 0, 1, -1]

    for a in range(4):
        new_i = i + dx[a]
        new_j = j + dy[a]
        if new_i >= 0 and new_i < 5 and new_j >= 0 and new_j < 5 and [new_i, new_j] not in new_visit:
            BFS(new_i, new_j, count_all, count_a, new_visit)


answer = 0
for i in range(5):
    for j in range(5):
        if map_[i][j] == 'A':
            BFS(i, j, 0, 0, [])
```
