# 소수구하기

```python
자연수 n이 주어질때 n을 서로 다른 두 소수의 곱으로 나타낼 수 있는지 알아보려 한다.
예를 들어, n =6 이면 2와 3의 곱으로 나타낼 수 있음. 그러나 n = 12면 안되겠지.
자연수 n이 주어졌을떄 두 소수의 곱으로 표현할수 있는지 찾아라

제한사항
1 n은 2이상 10^12 이하 자연수
2 두 소수를 오름차순으로 정렬해서 배열형태로 리턴
3 만약 나타낼수 없을시 [-1,-1] 로 리턴

ex)
n = 6   |       result = [2,3]
n =12   |       result = [-1,-1]
```

```python
from collections import deque
def solution(n):
    k = 2
    deq = deque()
    
    while n!= 1:
        if n%k == 0:
            deq.append(k)
            n/= k 
            if len(deq) >= 3:
                return [-1,-1]
        
        else:
            k+=1
    
    if len(set(deq)) != 2:
        return [-1,-1]
    
    else:
        return list(deq)
```

```python
class CircularQueue:
    def __init__(self,max_qsize):
        self.front = 0
        self.rear  = 0
        self.items = [None] * max_qsize
    
    # front ,rear pointer 위치 같으면 empty
    
    def isEmpty(self):
        return self.front == self.rear
    
    def isFull(self):
        return self.front == (self.rear + 1) % max_qsize
    
    def clear(self):
        self.front = self.rear # 이게 왜클리어
    
    def __len__(self):
        return (self.rear - self.front + max_qsize) % max_qsize
    
    def Push(self, item):
        if not self.isFull():
            self.rear = (self.rear + 1) % max_qsize
            self.items[self.rear] = item
        else:
            print("Overflow")
    
    def Pop(self):
        if not self.isEmpty(): 
            self.front = (self.front + 1 ) % max_qsize    
            return self.items[self.front]
        else:
            print("Underflow")
    
    def Front(self):
        if not self.isEmpty():
            print(self.items[(self.front+1)%max_qsize])
        else:
            print('NULL')
    
            
max_qsize, command = list(map(int,input().split()))
max_qsize= max_qsize +1
q= CircularQueue(max_qsize )
for _ in range(command):
    string = input()
    token = string.split()
    if token[0] =='1':
        q.Push(token[1])
    elif token[0] == '2':
        q.Pop()
    elif token[0]== '3':
        q.Front()
```

```python
from collections import defaultdict

numbers_of_nodes , node_a ,node_b = list(map(int , input().split() ))
a,b = node_a , node_b
map_ = defaultdict(dict)
for _ in range(numbers_of_nodes-1):
    up, down  = list( map(int , input().split() ))
    if up not in map_:
        map_[up] = {'up':-1}
    if down not in map_:
        map_[down] = {'up':-1}
    map_[down]['up'] = up

a_ancestor= []
b_ancestor = []

while map_[node_a]['up'] != -1:
    up_a = map_[node_a]['up']
    a_ancestor=   a_ancestor + [node_a]
    node_a = up_a

while map_[node_b]['up'] != -1:
    up_b = map_[node_b]['up']
    b_ancestor=   b_ancestor + [node_b]
    node_b = up_b
a_ancestor.append(0)
b_ancestor.append(0)

common  = list(set(a_ancestor) & set(b_ancestor))
min_ = len(a_ancestor)
for i in common:
    loc = a_ancestor.index(i)
    min_ = min(min_,loc)
common_ancestor = a_ancestor[min_]

count_a = 0
while a != common_ancestor: 
    a = map_[a]['up']
    count_a+=1
    
count_b = 0
while b != common_ancestor: 
    b = map_[b]['up']
    count_b+=1
    
print(count_a + count_b)
```

# 이분 그래프

```python

      1
      2
    4 
  6   3  5
  
  
  # 일반적인 트리가 아닐 경우 이분이 안됨 
     1
  2       3  4   # 여기서 2랑 4가 연결, 3이랑 4가 연결되잇음
```

```python
스택에다 경로를 추가할때 visited에 있는애를 추가해야하는 경우가 생긴다면 이분그래프임.
visited 가 tRUE 인데 또 VISITED에 갈수있다면 이분그래프?
```

```python
def solution():
    map_ = {}
    __ , command_times = list(map(int,input().split()))
    for _ in range(command_times):
        node_a, node_b = list(map(int,input().split()))
        if node_a not in map_:
            map_[node_a] =[]
        if node_b not in map_:
            map_[node_b] =[]

        map_[node_a].append(node_b)
        map_[node_b].append(node_a)
    
    visited= [False for _ in range(__ +1)]
    visited[0] =True
    
    q = []
    while sum(visited) != (__+1):
        for v in range(len(visited)):
            if visited[v] == False:
                q.append(v)
                break
        while q:
            go  = q.pop()
            visited[go] =True
            next_nodes = map_[go]
            # 방문 안한거만 q에 추가
            next_node_no_visit = []
            for nn in next_nodes:
                if visited[nn] ==False:
                    next_node_no_visit.append(nn)
            print(next_node_no_visit,'next_node_no_visit')
            print(q,'q')
            for nn in next_node_no_visit:
                if nn in q:
                    print('No')
                    return 
            q.extend(next_node_no_visit)
    
    print('Yes')
    
        
```
